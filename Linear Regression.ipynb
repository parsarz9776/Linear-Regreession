```python
# Importing the required libraries 
import numpy
import matplotlib.pyplot as plot
import pandas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# import Dataset
dataset1=pandas.read_excel(r'C:\Users\ParsaRz\Desktop\assignment 2 ML\parkinglot1.xlsx')
print(dataset1)
dataset2=pandas.read_excel(r'C:\Users\ParsaRz\Desktop\assignment 2 ML\parkinglot2.xlsx')
print(dataset2)
dataset3=pandas.read_excel(r'C:\Users\ParsaRz\Desktop\assignment 2 ML\parkinglot03.xlsx')
print(dataset3)
dataset4=pandas.read_excel(r'C:\Users\ParsaRz\Desktop\assignment 2 ML\parkinglot04.xlsx')
print(dataset4)
dataset5=pandas.read_excel(r'C:\Users\ParsaRz\Desktop\assignment 2 ML\parkinglot05.xlsx')
print(dataset5)
```

```python
# dedicating x to days and y to occupancy
x1=dataset1.iloc[:,:1].values
y1=dataset1.iloc[:,-1].values

x2=dataset2.iloc[:,:1].values
y2=dataset2.iloc[:,-1].values

x3=dataset3.iloc[:,:1].values
y3=dataset3.iloc[:,-1].values

x4=dataset4.iloc[:,:1].values
y4=dataset4.iloc[:,-1].values

x5=dataset5.iloc[:,:1].values
y5=dataset5.iloc[:,-1].values
```

```python
# Splitting the dataset into train and test 
x1_Train , x1_Test , y1_Train , y1_Test = train_test_split(x1,y1,test_size=1/3 , random_state=0)
x2_Train , x2_Test , y2_Train , y2_Test = train_test_split(x2,y2,test_size=1/3 , random_state=0)
x3_Train , x3_Test , y3_Train , y3_Test = train_test_split(x3,y3,test_size=1/3 , random_state=0)
x4_Train , x4_Test , y4_Train , y4_Test = train_test_split(x4,y4,test_size=1/3 , random_state=0)
x5_Train , x5_Test , y5_Train , y5_Test = train_test_split(x5,y5,test_size=1/3 , random_state=0)
```

```python
# Creating the Regression model for all parkings
LinearRegression1=LinearRegression()
LinearRegression1.fit(x1_Train,y1_Train)

LinearRegression2=LinearRegression()
LinearRegression2.fit(x2_Train,y2_Train)

LinearRegression3=LinearRegression()
LinearRegression3.fit(x3_Train,y3_Train)

LinearRegression4=LinearRegression()
LinearRegression4.fit(x4_Train,y4_Train)

LinearRegression5=LinearRegression()
LinearRegression5.fit(x5_Train,y5_Train)
```

```python
# Creating the prediction usuing test data
yPrediction1= LinearRegression1.predict(x1_Test)
yPrediction2= LinearRegression2.predict(x2_Test)
yPrediction3= LinearRegression3.predict(x3_Test)
yPrediction4= LinearRegression4.predict(x4_Test)
yPrediction5= LinearRegression5.predict(x5_Test)
```

```python
#plot for parking lot 1
plot.figure(figsize=(20,10))
plot.scatter(x1_Test,y1_Test , color = 'blue')
plot.plot(x1_Test, yPrediction1 , color='green')
plot.title('occupancy of parking 1')
plot.xlabel('Days')
plot.ylabel('Occupancy (%)')
```
